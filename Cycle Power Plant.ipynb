#### iii. mean, the median, range, first and third quartiles, and interquartile ranges

```python
## for Every Variable--> Temperature, Exhaust Vacuum, Ambient Pressure , Relative Humidity, Electrical Energy Output
##Stats function calculate mean,median,min,max--> to compute range,1st and 3rd Quantile -->with that IQR as well

def stats(var):
    mean = CPP_df[var].mean()
    median=CPP_df[var].median()
    max=CPP_df[var].max()
    min=CPP_df[var].min()
    range=max-min
    ## from documentation https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.DataFrame.quantile.html
    Q1 = CPP_df[var].quantile(0.25)
    Q3 = CPP_df[var].quantile(0.75)
    IQR= Q3-Q1
    return { 'mean' : mean, 'median':median, 'range':range,'Q1':Q1, 'Q3':Q3,'IQR':IQR}
```

```python
##Using Cubic Polynomial Regression --> we need cubic equation
### checking if the relationship between X and Y is nonlinear

##Temperature Vs EEO
##Using the Cubic Equation Given
# Temp_X = data['Temperature']
# X_poly = np.column_stack([Temp_X, Temp_X**2, Temp_X**3])
# X_poly = sm.add_constant(X_poly)
# ##Using Stats Model(suggested on Piazza)
# Temp_model_NL = sm.OLS(data['Electrical Energy Output'], X_poly).fit()
# # Print the model summary
# print(model.summary())
# hypotheses = 'x2 = 0, x3 = 0'
# f_test = model.f_test(hypotheses)
# print(f_test)
##Tried doing one at a time howver it is easir to make one function and send all prediators and compute non-linearity

Temperature=CPP_df['Temperature']
Exhaust_Vaccume=CPP_df['Exhaust Vacuum']
Ambient_Pressure=CPP_df['Ambient Pressure']
Relative_Humidity=CPP_df['Relative Humidity']
output_label=CPP_df['Electrical Energy Output']

##Using Scikit-learn->API Reference->sklearn.preprocessing->PolynomialFeatures reference
def implementing_cubic_equation(x, y, selected_predictor):
    print(f"Analyzing Cubic Regression for: {selected_predictor}")

    #Creating cubic polynomial features (X, X^2, X^3)
    X_cubiceq = [[1, xi, xi**2, xi**3] for xi in x]

    ###I tired to do it with my already created model above but it gave me error so we have to fit the model again using non_linear variables
    nonLinear_model = sm.OLS(y, X_cubiceq).fit()

    print(nonLinear_model.summary())

    #Performing F-test for X² and X³ terms ref:https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.f_test.html
    f_test = nonLinear_model.f_test("x2 = 0, x3 = 0")
    print("\nF-test for Non-Linearity:")
    print(f_test)

    if f_test.pvalue < 0.05:
        print(f"{selected_predictor} has significant non-linear components.")
    else:
        print(f"{selected_predictor} does NOT show strong non-linearity.")

    # ##ref: https://matplotlib.org/stable/gallery/lines_bars_and_markers/simple_plot.html
    x_sorted = sorted(x) ##line looked incorrect without sorting it was showing like a plane not a single line I don't know why(suggested by GPT to sort)
    y_pred = [nonLinear_model.predict([1, xi, xi**2, xi**3])[0] for xi in x_sorted]

    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color='teal', label='Actual Data')
    plt.plot(x_sorted, y_pred, color='pink', label='Cubic Fit')
    plt.xlabel(selected_predictor)
    plt.ylabel("Output Label")
    plt.title(f"Non linear Regression: {selected_predictor} vs Output Label")
    plt.legend()
    plt.show()

# Run cubic regression for each predictor
implementing_cubic_equation(Temperature, output_label, "Temperature")
implementing_cubic_equation(Exhaust_Vaccume, output_label, "Exhuast Vaccum")
implementing_cubic_equation(Ambient_Pressure, output_label, "Ambient Pressure")
implementing_cubic_equation(Relative_Humidity,output_label,"Relative Humidity")
```

$$
